```python
import numpy as np
import pandas as pd
import seaborn as sns

from matplotlib import pyplot as plt
%matplotlib inline
sns.set_style("whitegrid")

```

```python
# loading the data
dataframe = pd.read_csv("Resources/songs.csv")
dataframe.head()
```

```python
# the columns in the dataframe
print(dataframe.keys())
```

```python
# checking for the null value in the data
pd.isnull(dataframe).sum()
```

```python
# dropping the non-valuable column
columns = [ 'track_album_id', 'track_album_name',
                  'track_album_release_date', 'playlist_name', 'playlist_id','track_id','playlist_subgenre','track_name','track_artist']
dropped = dataframe.drop(columns=columns)
dropped
```

```python
# dropping the null values 
df_genre_clean = dropped.dropna(subset='playlist_genre')
df_genre_clean = df_genre_clean.dropna(how='any')
```

```python
# confirming that all null values are dropped
pd.isnull(df_genre_clean).sum()
```

```python
sns.distplot(df_genre_clean['track_popularity']).set_title('Popularity Distribution')
```

```python
ax = df_genre_clean['track_popularity'].plot.hist(density= True, bins = 10, color = 'c', alpha = 0.6)
df_genre_clean['track_popularity'].plot.density(ax=ax)
```

```python
# determining the correlation between the dataset
df_genre_clean.corr()
```

```python
# visualizing the correlation between our dataset
sns.heatmap(df_genre_clean.corr())
```

```python
# setting our target value to popular and non popular based on track popularity

df_genre_clean.loc[df_genre_clean['track_popularity'] < 55, 'track_popularity'] = 0
df_genre_clean.loc[df_genre_clean['track_popularity'] >= 55, 'track_popularity'] = 1
df_genre_clean.loc[df_genre_clean['track_popularity'] == 1]
```

```python
# changing the non-numeric value to a numeric value
new_dummy=pd.get_dummies(df_genre_clean['playlist_genre'])
new_dummy
```

```python
# dropping the non numeric value
df_genre_clean=pd.concat([df_genre_clean, new_dummy], axis=1)
df_genre_clean=df_genre_clean.drop(columns=['playlist_genre'],axis=1)
df_genre_clean
```
