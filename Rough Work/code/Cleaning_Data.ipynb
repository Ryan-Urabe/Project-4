```python
copy = drop_subg.copy()
```

```python
copy
```

```python
for i in columns:
    quartiles = drop_subg[i].quantile([.25, .5, .75])
    lower = quartiles[0.25]
    upper = quartiles[0.75]
    iqr = upper-lower
    
    lower_bound = lower - (1.5*iqr)
    upper_bound = upper + (1.5*iqr)
    x = copy[i]
    for info in x.index:
        if x[info] < lower_bound or x[info] > upper_bound:
            copy.drop(info, inplace=True)
```
